```python
book_title = []
author = []
read_status = []
authors_name = []
total_read_books = 0
total_unread_books = 0

number_request = int(input('enter number of your request:')) 

for i in range(number_request):
    request = input('enter your request or help:')
    if request == 'add':
        title = input('enter book title:')
        if title in book_title:
            print(f'{title} title already exist')

        else:
            number_authors = int(input('enter number authors for this title'))  
            for i in range(number_authors):
                authors = input('enter authors name')
                authors_name.append(authors)
            author.append(authors_name)
            authors_name = []
            book_title.append(title)   
            read = input('if u read this title enter true else false?')
            read_status.append(True) if read == 'true'  else read_status.append(False)

    elif request == 'search':
        ask = input('what would you like to search? title or author !')
        if ask == 'title' :
             title = input('enter your title search?')
             if title in book_title:
                idx = book_title.index(title)
                print(f'information book title {title} is : read status is {read_status[idx]} and author is {author[idx]}')
             else:
                print(f'this {title} not found.')

        if ask == 'author' :
            author_name = input('enter autor name?')
            flag = False
            for i in range(len(author)):
               if author_name in author[i]:
                   flag = True
                   print(f'{author_name} is author for {book_title[i]}.')
            
            if flag == False :
               print(f'{author_name} not found . ')

    elif request == 'show':
        ask = input('what would you like to show? all or read or unread')
        if ask == 'all':
           for i in book_title :
             print(f'book {i}')

        elif ask == 'read':
           for i in range(len(read_status)):
             if read_status[i] == True:
                print(f'read books : {book_title[i]}')
                total_read_books += 1
        elif ask == 'unread': 
               for i in range(len(read_status)):
                   if read_status[i] == False:               
                       print(f'unread books: {book_title[i]}')
                       total_unread_books += 1

    elif request == 'summary':
        total_books = len(book_title)
        print(f'the total books are {total_books} and the total read books are {total_read_books} and the total unread books are {total_unread_books}')         

    elif request == 'help':
        print('add --------> adding new book.')
        print('search ------> by title or by author')                         
        print('show --------->show the books and read/unread books and total books/read/unread books.')
        print('summary --------> show all books/read books/unread books')
        print('exit ----------> for existing the program')

    elif request =='exit':
        break

    else:
        print('command not found.')  


print(book_title,'\n',author,'\n',read_status)
```
